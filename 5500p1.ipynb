```python
import pandas as pd

# Specify the path to your data file
file_path = '/content/tracking-the-t/data_accuracy/Rapid_Transit_and_Bus_Prediction_Accuracy_Data.csv'

# Load the data into a DataFrame
df = pd.read_csv(file_path)

df
```

```python
df['num_missed_predictions'] = df['num_predictions']-df['num_accurate_predictions']
df['missed_percent'] = df['num_missed_predictions'] / df['num_predictions']
df['accurate_percent'] = df['num_accurate_predictions'] / df['num_predictions']
```

```python
grouped_average = df.groupby('mode')['accurate_percent'].mean().reset_index()

print(grouped_average)
```

```python
grouped_average_bin = df.groupby('bin')['accurate_percent'].mean().reset_index()

print(grouped_average)
```
